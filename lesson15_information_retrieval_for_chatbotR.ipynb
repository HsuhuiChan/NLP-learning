# Boolean search

+ X是我们的tfidf向量矩阵。
+ 每一行代表的是该文章中每个单词的tfidf值。
+ 矩阵转置之后，每一行代表该单词在文章中出现的情况。

```python
transposed_x = X.transpose().toarray()
```

```python
word_2_id = vectorized.vocabulary_
```

基于向量的搜索引擎

```python
def search_engine(query):
    """
    @query is the search words, splited by space
    @return is the related ducuments which ranked by tfidf similarity
    """
    words = query.split()
    query_vectory = vectorized.transform([' '.join(words)]).toarray()[0]
    words_ids = [word_2_id[w] for w in words]
    documents_ids = [set(np.where(transposed_x[id_])[0]) for id_ in words_ids]
    merged_documents = reduce(and_, documents_ids)
    # we could know the documents which contain these words
    sorted_documents_ids = sorted(merged_documents, key=lambda i:distance(query_vectory, X[i].toarray()))
    return sorted_documents_ids
```

```python
%%time
search_engine('美军 司令 航母')
```

```python
def get_query_pat(query):
    return re.compile('({})'.format('|'.join(query.split())))
```

```python
get_query_pat('美军 司令 航母')
```

```python
def highlight_keywords(pat, document):
    return pat.sub(repl='***\g<1>***', string=document)
```

```python
highlight_keywords(pat=get_query_pat('清华 大学'), document=content['content'][11430])
```

```python
def search_engine_with_pretty_print(query):
    candidates_ids = search_engine(query=query)
    for i, _id in enumerate(candidates_ids):
        title = '## Search Result {}'.format(i+1)
        c = news_content[_id]
        c = highlight_keywords(pat=get_query_pat(query=query),document=c)
        display(Markdown(title + '\n' + c))
```
